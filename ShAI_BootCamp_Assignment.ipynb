
# **Answering the 4th question**

 Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.


```python
New.hist(column='TotalPay')
```

```python
# use pie charts to represent the proportion of employees in different departments.
import re
x  = New.shape[0]  ##Numer of rows
dep= New['JobTitle']
z=list(range(x))
for i in range(x):
  z[i]= re.findall(r'\(.*?\)',dep[i])
print(z)
```

```python
#Remove empty element from Z
result = list(filter(None, z))
print(result)
```

```python
#use pie charts to represent the proportion of employees in different departments.
flat_list = [item for sublist in result for item in sublist]
counts = {}
for item in flat_list:
    if item not in counts:
        counts[item] = 0
    counts[item] += 1
for item, count in counts.items():
    print(f"{item}: {count}")
```

```python
import matplotlib.pyplot as plt
labels = list(counts.keys())
sizes = list(counts.values())
fig, ax = plt.subplots()
texts, autotexts = ax.pie(sizes, labels=labels)
```
